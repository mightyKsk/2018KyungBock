### 1_2) 10개의 데이터만 출력

```python
f=open('seoul_temp.csv','r')
import csv
data=csv.reader(f)
header=next(data) #정보를 알려주는 데이터 건너띄기
count=0
for line in data :
    count+=1
    if(count==10) : break
    print(line)
```

### 1_3) 최저 값을 찾는 알고리즘으로 최저 기온을 찾기

```python
f=open('seoul_temp.csv','r')
import csv
data=csv.reader(f)
header=next(data)

minTemp=0 #최저기온을 저장하기 위한 변수
minTempDate="" #최저기온이었던 날짜를 저장하기 위한 변수
for line in data :
    if (minTemp > float(line[-2])) : #최저기온 line[-2] , 문자데이터이므로 실수형으로 변환하여 최저값 찾기
        minTemp=float(line[-2])
        minTempDate=line[0]
    
print("서울이 가장 추웠던 날 : %s / %2.1f도"%(minTempDate ,minTemp))
date=minTempDate.split('-') 
year,month,day=date[0],date[1],date[2]

print("서울이 가장 추웠던 날은 %s년 %s월 %s일 입니다 : %2.1f도"%(year,month,day,minTemp))
```

## 2. 서울의 기온 분석(그래프)
### 1)2001년 01월 01일~2018년 11월 22일까지의 기온을 저장 : seoul_temp.csv
### 2) 서울의 평균 기온 그래프로 나타내기

### 2_1) 데이터를 읽어와서 평균기온 나타내기 

```python
f=open('seoul_temp.csv','r')
import csv
data=csv.reader(f)
header=next(data) #정보를 알려주는 데이터 건너띄기
count=0
for line in data :
    count+=1
    if(count==10) : break
    print(line[1])
```

### 2_2) 꺾은 선 그래프로 평균기온 표현하기 

```python
f=open('seoul_temp.csv','r')
import csv
import matplotlib.pyplot as plt #그래프를 그리기 위해 임포트
data=csv.reader(f)
header=next(data) #정보를 알려주는 데이터 건너띄기
result=[]
for line in data :
    if line[1]!='' :
        result.append(float(line[1]))

plt.plot(result)
plt.show()
```

#### 7월의 데이터만 빼오기

```python
f=open('seoul_temp.csv','r')
import csv
data=csv.reader(f)
header=next(data) #정보를 알려주는 데이터 건너띄기
summer=[]
for line in data :
    month=line[0].split('-')[1] #날짜 데이터를 '-'단위로 잘라서 2번째 값이 월
    if(line[1]!='' and month =='07') :
        summer.append(float(line[1]))

print(summer)
```

```python
f=open('seoul_temp.csv','r')
import csv
import matplotlib.pyplot as plt

data=csv.reader(f)
header=next(data) 
summer=[]
for line in data :
    month=line[0].split('-')[1] #월에 관한 정보는 날짜를 '-'단위로 자른 후 2번째 값
    if(line[1]!='' and month =='07') :
        summer.append(float(line[1]))
plt.style.use('ggplot')
plt.hist(summer, bins=100)
plt.show()
```
