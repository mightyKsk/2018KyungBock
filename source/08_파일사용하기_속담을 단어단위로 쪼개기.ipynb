### 1) 속담을 단어단위로 쪼개기

```python
f=open("proverbs.txt","r")
words=[]
while True :
    line=f.readline() #한 문장씩 읽기
    print(line)
    if(line=='') : break #문장이 끝나면 반복문 빠져나오기
    line=line.rstrip() #오른쪽 공백 없애기
    #문장을 단어 단위로 쪼개기    
    words+=line.split(' ')
    
print(words)

for word in words :
    if word.endswith('.') :
        word=word[:-1]
    print(word)
f.close()
```

###  2) 속담을 단어 단위로 쪼개고 나오는 빈도 계산 

```python
import string
f=open("proverbs.txt","r")

words=[]
while True :
    line=f.readline() #한 문장씩 읽기
    print(line)
    if(line=='') : break #문장이 끝나면 반복문 빠져나오기
    line=line.rstrip() #오른쪽 공백 없애기
    line=line.translate(line.maketrans("","",string.punctuation))
    line=line.lower()
    #문장을 단어 단위로 쪼개기    
    words+=line.split(' ')
    
proverbs={}
for word in words :
    if len(word)>3:
        if word in proverbs :
            proverbs[word]+=1
        else :
            proverbs[word]=1
        
import csv
of=open("myProverbs.csv","w",newline='')
w=csv.writer(of)

for key, value in proverbs.items() :
    print(key,":",value)
    w.writerow([key,value])

f.close()
of.close()
```
